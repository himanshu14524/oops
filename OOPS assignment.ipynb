```python
Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

ANS--Multiple inheritance is a feature in object-oriented programming
languages where a class can inherit attributes and methods from more
than one parent class. This allows a child class to inherit and combine 
the properties and behaviors of multiple parent classes.

class Himanshu:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")


class coder:
    def code(self):
        print(f"{self.name} is a coder.")


class developer(Himanshu,coder):
    def __init__(self, name):
        super().__init__(name)


developer =developer("sachin")
developer.eat()    # Output: sachin is eating.
developer.sleep()  # Output: sachin is sleeping.
developer.code()    # Output: Sachin is a coder.
```

```python
Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

ANS--Method overriding is a feature in object-oriented programming where a subclass 
provides a different implementation for a method that is already defined in its
parent class. The subclass overrides the method to provide its own implementation
that is specific to its needs.

class Animal:
    def sound(self):
        print("Animal makes a sound.")


class Dog(Animal):
    def sound(self):
        print("Dog barks!")


class Cat(Animal):
    def sound(self):
        print("Cat meows!")


animal = Animal()
animal.sound()  # Output: Animal makes a sound.

dog = Dog()
dog.sound()     # Output: Dog barks!

cat = Cat()
cat.sound()     # Output: Cat meows!
 
```
